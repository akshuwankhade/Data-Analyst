# Loops
Loops allowed you to repeatedly execute a block of code, while and for loop

```python
# While loop
# repeatedly execute a block of code as long as condition true

n = 5
i = 1
while i <= n:
  print(i)
  i=i+1
```

```python
count = 5
while count > 0:
  print(count)
  count=count-1
```

```python
n = 7
i = 1
while i < n:
  print(i)
  i=i+1
  if i == 3:
    break
else:
  print("This will  be executed  when the  while loop is run succesfully")
```

```python
# Continue >> to skip  that iteration

n = 7
i = 1
while i < n:
  i=i+1
  if i == 3:
    continue
  print(i)
else:
  print("This will  be executed  when the  while loop is run succesfully")
```

```python
# For loop >> iterate over a squence if elements

a = "pwskills"
for char in a:
  print(char)
```

```python
l = [1,2,"Ajay","pwskills"]
for i in l:
  print(i)
```

```python
for i in l:
  print(i)
else:
  print("This will execute when for loop ends")
```

```python
for i in range(10):
  print(i)
```

```python
# Syntax range(Start, Stop, Step)
list(range(0,12,3))
```

```python
# Print a right triangle using for loop

for i in range(4):
  for j in range(i+1):
    print("*", end = "")
  print()
```

```python
number = [1,2,3,4,5,6,7]
for number in number:
  if (number == 3) or (number==7):
    continue
  print(number)
```

```python
number = [1,2,3,4,5,6,7]
for number in number:
  if (number == 3) or (number==7):
    break
  print(number)
```

```python
# print odd number
for i in range(10):
  if i % 2==0:
    continue
  print(i)
```

```python
# print even Number
for i in range(10):
  if i % 2 !=0:
    continue
  print(i)
```
